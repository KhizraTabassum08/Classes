```python
#Task 1:
class Person:
    def __init__(self,name,age,city):
         self.name = name
         self.age = age
         self.city = city
    def __str__(self):
      return (f"Name:{self.name}, Age: {self.age}, City: {self.city}")
my_person = Person("lisa",25,"busan")
print(my_person)
```

```python
#Task 2:
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    def __str__(self):
        return (f"{self.make},{self.model},{self.year}.")
my_car = Car("BMW","m5",2024)
print(my_car)
```

```python
#Task 3:
class Circle:
    def __init__(self,radius):
      self.radius = radius

    def calculate_area(self):
        return 3.1415*(self.radius**2)
    def calculate_circumference(self):
       return 2*3.1415*self.radius
    def __str__(self):
        return (f"circle with radius{self.radius}.")
my_circle = Circle(6)
print(my_circle)
print(f"Area:{my_circle.calculate_area(): .2f}")
print(f"Circumference: {my_circle.calculate_circumference(): .2f}")
```

```python
#Task 4:
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
       return self.length + self.width
    def perimeter(self):
       return 2*(self.length + self.width)
    def __str__(self):
       return (f"rectangle with length:{self.length} and width:{self.width}")
my_rectangle = Rectangle(6,8)
print(my_rectangle)
print(f"Area:{my_rectangle.area(): .2f}")
print(f"Perimeter:{my_rectangle.perimeter(): .2f}")
```

```python
#Task 5:
class Book:
    def __init__(self,title,author,publication_year):
        self.title = title
        self.author = author
        self.publication_year = publication_year
    def __str__(self):
        return (f"Title:{self.title}, Author: {self.author}, Year: {self.publication_year}")
my_book = Book("Abe Hayat","Umera Ahmed",2017)
print(my_book)
```

```python
#Task 6:
import math 
class Shape:
    def area(self):
        raise notimplementedError("Subclasses should implement this method!")
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * (self.radius **2)
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
class Triangle(Shape):
    def __init__(self, base,height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base *self.height
circle = Circle(5)
rectangle = Rectangle(4,6)
triangle = Triangle(3,7)
print(f"Circle area: {circle.area(): .2f}")
print(f"Rectangle area: {rectangle.area()}")
print(f"Triangle area: {triangle.area()}")
```
